## The effect of temperature on the amount of rainfall in Mumbai

The water cycle is a representation of how rainfall occurs. Water on the earth’s surface evaporates and rises until it reaches a height where there are extremely low temperatures. Due to the low temperature, the water vapour condenses to form water droplets on tiny dust particles. These particles accumulate to form clouds. When the amount of water droplets starts to increase the cloud gets heavier making it difficult for the water to remain suspended in the air. After one point the clouds become too heavy and the droplets fall down as rain.

A key factor to forming clouds is the amount of water that evaporates off the earth’s surface. A faster rate of evaporation occurs at higher temperatures because water is provided with more heat energy to convert from a liquid to gas.

When directly comparing temperature and rainfall the relation is not exactly directly proportional. When looking at weather patterns the trend that is visible is that after a time span of high temperatures where the water evaporates rainfall occurs but the days when rainfall occurs the temperatures are relatively lower than before the rain actually fell. 

Besides temperature, the wind is also a factor when considering the amount of rainfall. The wind blows over water bodies and collects moisture that in turn helps form rain in the form of clouds. The stronger winds the more rainfall a place will experience. 

Mumbai is a small coastal city which has moderately hot temperatures all year round resulting in heavy rains throughout the year as well. Summer in Mumbai starts in the month of April, continues through May, the hottest period, and then ends early June. Summer is followed by monsoon at the beginning of the month of June. Clearly visible in the trend displayed right after the hottest time period of the year, rainfall starts to begin and in a coastal city like Mumbai where there is high humidity and huge water bodies surrounding the city, rainfall is heavy. Location is an important factor when rainfall is being measured. Locations with water bodies nearby will experience greater rainfall compared to a landlocked location because landlocked areas will not have water bodies nearby to evaporate and form clouds. For example, Mumbai experiences more rainfall overall compared to New Delhi which is landlocked. 


#### The following code uses polyfit from numpy to find a relation between the temperature and rainfall over a years period in Mumbai.

Importing libraries

```python
import numpy
import matplotlib.pyplot as plt
```

Reading through the data set and creating a list using the temperature as x values and rainfall as y values



```python
x_file = open('data_x.txt', 'r')
y_file = open('data_y.txt', 'r')
x=[]
y=[]
for i in x_file:
    val=i.strip('\n')
    x.append(int(val))
    
for i in y_file:
    val=i.strip('\n')
    y.append(float(val))
```

Used matplotlib to plot a scatter plot using the x and y values.

```python
plt.scatter(x, y)
plt.xlabel('Temperature /degrees Celcius')
plt.ylabel('Rainfall /mm')
plt.show()
```

Used polyfit to fit a line of best fit for the given data by trying different degrees but the line of best fit never fit the data properlly due to the multiple y values for each x value.

```python
mymodel = numpy.poly1d(numpy.polyfit(x, y, 4))
myline = numpy.linspace(24, 34, 100)
plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.xlabel('Temperature /degrees Celcius')
plt.ylabel('Rainfall /mm')
plt.show()
```

Found the average amount of rainfall for each temperature to help fit a better line of best fit.

```python
temp=24
x_mean=[]
for i in range(24,34):
    x_mean.append(i)
y_mean=[]
while temp<34:
    c=1
    total=0
    for i in range (0,360):
        if x[i]==temp and y[i]>0:
            total+=y[i]
            c+=1
        else:
            pass
    mean=total/c
    y_mean.append(mean)
    temp+=1 
```

Plot a scatter plot using a mean values calculated.

```python
plt.scatter(x_mean,y_mean)
plt.xlabel('Temperature /degrees Celcius')
plt.ylabel('Average rainfall /mm')
plt.show()
```

Used polyfit again to fit a line of best fit.

```python
mymodel = numpy.poly1d(numpy.polyfit(x_mean, y_mean, 4))
myline = numpy.linspace(24, 34, 100)
plt.scatter(x_mean, y_mean)
plt.xlabel('Temperature /degrees Celcius')
plt.ylabel('Average rainfall /mm')
plt.plot(myline, mymodel(myline))
plt.show()
```
